##### 下載儲存圖表
使用 matplotlib 繪製圖表之後，可以透過 savefig() 方法下載儲存圖表，這篇教學會介紹相關的方法。

##### savefig() 常用參數 



##### 儲存圖表到 Google 雲端硬碟 
與 Google Drive 連動後，執行下方的程式碼，就能將圖表以透明背景的 png 格式，儲存到指定的目錄裡 ( 預覽時會以灰色方格表示透明背景 )。

* [參考：連動 Google Drive](https://steam.oxxostudio.tw/category/python/info/online-editor.html#a4)



```python
from matplotlib import pyplot as plt

x = [1,2,3,4,5]
fig = plt.figure()
plt.plot(x)
plt.title('test', fontsize=20)
plt.xlabel('X', fontsize=20)
plt.ylabel('Y', fontsize=20)
plt.savefig('/content/drive/MyDrive/Colab Notebooks/test.png',
            transparent=True,
            bbox_inches='tight',
            pad_inches=1)
plt.show()
```
